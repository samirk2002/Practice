Reverse array

```python
arr=[1,2,3,4,5,5,6]
start=0
end= len(arr)-1

while start <end :
    arr[start],arr[end]=arr[end],arr[start]
    start+=1
    end-=1

print(arr)
```

```python
arr=[1,2,3,4,5,6,6,7,8]
rev_arr=[]

for i in range(len(arr)-1,-1,-1):
    rev_arr.append(arr[i])
print(rev_arr)
```

```python
arr=[6,5,4,3,2,1]
stack=[]

for i in arr:
    stack.append(i)

for i in range(len(arr)):
    arr[i]=stack.pop()

print(arr)
```

```python
arr=[1,2,3,4,5,67,8]
rev_arr2=arr[::-1]
print("Using Slicing :",rev_arr2)

rev_arr=list(reversed(arr))
print("Using Reversed Function : ",rev_arr)
```

Maximum element

```python
list1=[1,2,3,5,54,7,7]

max_val=list1[0]

for i in list1:
    if i> max_val:
        max_val=i
print("Maximum : ", max_val)
```

```python
#Bubble sort

list1=[1,2,3,5,54,7,7]

for i in range(len(list1)):
    for j in range(0,len(list1)-i-1):
        if list1[j]> list1[j+1]:
            list1[j],list1[j+1]=list1[j+1],list1[j]

print("maximum : ",list1[-1])
```

```python
def find_max(arr,n):
    if n==1:
        return arr[0]
    return max(arr[n-1], find_max(arr,n-1))
arr=[1,2,3,4,5,6,7]
result= find_max(arr,len(arr))

print("Maximum : ", result)
```

```python
from functools import reduce
arr=[1,2,3,4,56,78]
maximum = reduce(lambda a,b:a if a> b else b, arr)
print(maximum)
```

second largest element

```python
arr=[10,20,40,99]
first=second=float('-inf')

for num in arr:
    if num > first:
        second=first
        first=num
    elif num > second and num != first:
        second = num

print("Second Largest : ", second)
```

```python
arr=[10,20,40,99]

for i in range(len(arr)):
    for j in range(0,len(arr)-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
largest=arr[-1]
for i in range(len(arr)-2,-1,-1):
    if arr[i]!= largest:
        print("Second Largest: ", arr[i])
        break
```

```python
arr=[10,20,40,99]

unique=list(set(arr))
unique.sort()
print("second Largest:", unique[-2])
```

```python
def second_large(arr):
    if len(arr)<2:
        return None
    largest=max(arr)
    arr.remove(largest)
    return max(arr)
arr=[10,20,40,99]
print("Second Largest: ", second_large(arr))
```

```python
#selection sort

arr=[10,20,40,99,5,25,55]
for i in range(len(arr)):
    min_index=i
    for j in range(i+1,len(arr)):
        if arr[j]<arr[min_index]:
            min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]
print("Sorted Array : ", arr)
```
